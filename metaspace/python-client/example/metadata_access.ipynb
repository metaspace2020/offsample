```python
%matplotlib inline
from metaspace.sm_annotation_utils import SMInstance
import matplotlib.pyplot as plt
```

This notebook provides a minimal example of accessing metadata for datasets submitted to the [METASPACE annotation platform](annotate.metaspace2020.eu)

```python
# Provide a path to the main configuration file used by SM engine
sm = SMInstance()
datasets = sm.datasets()
datasets = datasets[0:3]
```

```python
# JSON schema can be found at https://github.com/metaspace2020/metaspace/tree/master/metaspace/metadata
metadata = sm.get_metadata()
metadata.columns # available metadata
```

```python
# Plot a pie chart of ionisation sources used
metadata['MS_Analysis.Ionisation_Source'].value_counts().plot(kind='pie', figsize=(5, 5))
```

```python
# Plot matrix popularity for MALDI datasets
maldi_metadata =  metadata.loc[metadata['MS_Analysis.Ionisation_Source']=='MALDI']

ax = maldi_metadata['Sample_Preparation.MALDI_Matrix'].value_counts()\
        .plot(kind='bar', figsize=(5, 5), title='MALDI matrices used', )
    
ax.set_ylabel('number of datasets')
plt.show()
```
